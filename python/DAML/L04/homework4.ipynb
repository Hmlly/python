# 2 证明 (3) 式的极⼩值点为
$$
w=(X^TX+\lambda I)^{-1}X^TY$
$$
$$
min_\omega=L(\omega)+\lambda||\omega||^2 \tag{3}
$$
证明：
$$
L(\omega)+\lambda||\omega||^2 = \frac{1}{m}(\omega^T X^T X \omega - Y^T X \omega - \omega^T X^T Y -Y^T Y) + \lambda \omega^T \omega
$$
令 $\frac{\partial}{\partial\omega}(L(\omega)+\lambda||\omega||^2) = 0$，则 $L(\omega)+\lambda||\omega||^2$最小。

$$
\begin{align}
&=> \frac{\partial}{\partial\omega}(L(\omega)+\lambda||\omega||^2) = 0\\
&=> \frac{\partial}{\partial\omega}(\frac{1}{m}(\omega^T X^T X \omega - Y^T X \omega - \omega^T X^T Y -Y^T Y) + \lambda \omega^T \omega) = 0\\
&=> \frac{\partial}{\partial\omega}(\omega^T X^T X \omega) - \frac{\partial}{\partial\omega}(Y^T X \omega+\omega^T X^T Y) + 0 + \frac{\partial}{\partial\omega}(m\lambda \omega^T \omega) = 0\\
&=> (\omega^T X^T X + (X^T X \omega)^T) - (Y^T X + (X^T Y)^T) + (m\lambda \omega^T + (m\lambda \omega)^T) = 0\\
&=> \omega^T(X^T X + m\lambda I) = Y^T X\\
&=> (X^T X + m\lambda I)\omega = X^T Y\\
&=> \omega =(X^T X + \hat{\lambda}I)^{-1} X^T Y \quad (\hat{\lambda}=m\lambda)
\end{align}
$$

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import mean_squared_error
sns.set()
````

````python
# 股票数据
price = pd.read_csv("../nyse/prices.csv")
price_GE = price.query("symbol == 'GE'") 
# 数据归一化
price_GE.eval('`volume` = `volume`/5000000',inplace=True)
columns = ['date','symbol','high','low','open','close','volume']
price_GE = price_GE.reindex(columns,axis=1)
price_GE.reset_index(drop=True,inplace=True)
# print(price_GE.shape) # (1762,7)
# price_GE.head(n=10)
````

````python
# 构建 X,Y
def build_X(price_GE,step_k):
    # 构建 X
    row_num = price_GE.shape[0]
    price_new = []
    price_new.append(price_GE)
    new_columns = price_GE.columns.tolist()
    for i in range(1,step_k+1):
        add_columns = ['high-{}'.format(i),'low-{}'.format(i),'open-{}'.format(i),'close-{}'.format(i),'volume-{}'.format(i)]
        # print(add_columns)
        price_last_ith =price_GE[0:row_num-i][columns[2:]]
        for j in range(i):
            price_last_ith = pd.concat([price_GE[0:1][columns[2:]],price_last_ith],axis=0)
        price_last_ith.columns = add_columns
        price_last_ith.reset_index(drop=True,inplace=True)
        assert(price_new[i-1].shape[0] == price_last_ith.shape[0])
        price_new.append(pd.concat([price_new[i-1],price_last_ith],axis=1))
    # price_new[i].head(n=7)
    return price_new[step_k]

    # 构建Y
def build_Y(price_GE,step_k):
    price_y = price_GE[step_k:]
    for i in range(step_k):
        price_y = pd.concat([price_y,price_GE[-1:]],axis=0)
    price_y.reset_index(drop=True,inplace=True)
    column_y = 'open+{}'.format(5)
    price_y.rename(columns={'open':column_y},inplace=True)
    # price_y.head(n=10)
    return price_y
````

````python
# 构建 MA EMA
def build_MA(price_GE,MA_N):
    price_MA = price_GE
    price_MA['MA']=price_GE['close'] # N日MA=N日收市价的总和/N(即算术平均数)
    price_MA['EMA']=price_GE['close'] # 首次上期EXPMA值为上一期收盘价，N为天数。
    for i in range(1,row_num):
        # N日MA=N日收市价的总和/N(即算术平均数)
        price_MA.loc[i,'MA'] = price_MA.loc[i-MA_N+1:i,'close'].mean() # .loc 行为闭区间
        # EXPMA=（当日或当期收盘价－上一日或上期EXPMA）/N+上一日或上期EXPMA
        price_MA.loc[i,'EMA'] = (price_MA.loc[i,'close']- price_MA.loc[i-1,'EMA'])/MA_N+price_MA.loc[i-1,'EMA']
    # price_MA.head()
    return price_MA
````

````python
# 构建特征集
def build_feature(price_GE,step_k,MA_N):
    # 构建 X Y
    price_x = build_X(price_GE,step_k)
    price_y = build_Y(price_GE,step_k)
    # 构建 MA EMA
    price_MA = build_MA(price_GE,MA_N)
    
    # 合并 X(包括MA EMA) Y 
    assert(price_x.shape[0] == price_y.shape[0] == price_MA.shape[0])
    price_xy = pd.concat([price_x,price_MA[['MA','EMA']],price_y[column_y]],axis=1)
    # print(price_xy.shape) # (1762, 35)
    price_xy.head(n=7)

    # 删除无效行
    drop_front = np.arange(step_k)
    # print(drop_front)
    price_xy = price_xy.drop(drop_front,axis=0)
    drop_end = price_xy.shape[0] + drop_front
    # print(drop_end)
    price_xy = price_xy.drop(drop_end,axis=0)
    # print(price_xy.shape) # (1752, 35)
    # price_xy.head(n=7)
    return price_xy
````

````python
# 构建数据集
def build_data(price_GE,step_k,MA_N):
    # 构建特征集
    price_xy = build_feature(price_GE,step_k,MA_N)
    
    # 拆分训练集和测试集
    df_train = price_xy[price_xy["date"] < "2016"] # (1505,35)
    # print(df_train.shape)
    df_test = price_xy[price_xy["date"] >= "2016"] # (247,35)
    # print(df_test.shape)

    # 训练集 shuffle
    # df_train = df_train.sample(frac=1).reset_index(drop=True)
    # df_train.reset_index(drop=True,inplace=True)
    # df_train.head()
    return df_train,df_test
````

````python
# 训练 - 线性回归
def mytrain(df_train_train,lamb):
    # X Y
    row_num = df_train_train.shape[1]
    data_X = df_train_train.iloc[:,2:row_num-1] # .iloc为开区间
    data_X['1'] = 1
    # print(data_X.shape) # (1204, 33)
    data_Y = df_train_train.iloc[:,row_num-1]
    # print(data_Y.shape) # (1204,)
    # 单位矩阵
    col_num = data_X.shape[1]
    Imatrix = np.identity(col_num) 
    # print(Imatrix.shape) # (33, 33)
    # 计算 w
    data_XTX = np.matmul(data_X.T.values,data_X.values)
    # print(data_XTX.shape) # (33, 33)
    data_inv = np.linalg.inv(data_XTX + lamb*Imatrix)
    data_XTY = np.matmul(data_X.T.values,data_Y.values)
    # print(data_XTY.shape) # (33,)
    omega = np.matmul(data_inv,data_XTY)
    # print(omega.shape) # (33,)
    # print(omega)
    return omega
````

````python
# 验证 df_train_verify
def myverify(df_train_verify,omega):
    # X Y
    row_num = df_train_verify.shape[1]
    data_X = df_train_verify.iloc[:,2:row_num-1] # .iloc为开区间
    data_X['1'] = 1
    # print(data_X.shape) # (301, 33)
    data_Y = df_train_verify.iloc[:,row_num-1]
    # print(data_Y.shape) # (301,)
    # 计算均方误差 MSE
    pred_Y = np.matmul(data_X.values,omega)
    MSE = mean_squared_error(data_Y.values,pred_Y)
    MAE = sum([abs(a-b) for a,b in zip(data_Y.values,pred_Y)])/len(pred_Y)
    # print(MSE)
    return MSE,MAE
````

````python
# 训练
# 超参数
# step_k = 5
# lamb = 2.0
# MA_N = 5

# 拆分训练集的数据集和验证集
# row_num = df_train.shape[0]
omega_list = []
MSE_list = []
mse_mean_list = []
MAE_list = []
mae_mean_list = []
verify_batch = 5 # 交叉验证批次
batch_num = int(df_train.shape[0] / verify_batch)
# 调试超参数
for j in range(0,10):
    # 超参数
    step_k = 0
    # step_k = j
# [8.201779085973228e-11, 0.032307319068250635, 0.11932063842517315, 0.19962259553720546, 0.2852583682570309, 
#  0.3696457073494027, 0.45949686034312653, 0.5440061834151034, 0.6371209466033527, 0.7267945727972742]
# [6.645385781145781e-06, 0.12407388009629347, 0.2503231845810161, 0.32699229850143574, 0.39312728248549017, 
#  0.45616065160351466, 0.5101761662694306, 0.5606044422070744, 0.6113267881977726, 0.6567773014286707]
    MA_N = 8
    # MA_N = 1+j
# [7.331011287665575e-11, 8.201779085973228e-11, 7.330637046322339e-11, 6.942801098584904e-11, 6.827697846305785e-11, 
#  7.615668766717737e-11, 7.419250167080815e-11, 6.105759546077542e-11, 6.186117935127724e-11, 6.259841932618254e-11]
# [6.155736435318934e-06, 6.645385781145781e-06, 6.312424329742328e-06, 6.093655355349153e-06, 5.971171364254957e-06, 
#  6.579534213450699e-06, 6.528418787849451e-06, 5.60318748320654e-06, 5.764162554199241e-06, 5.752215557002672e-06]
    # lamb = 0.001
    lamb = pow(10,-5+j)
# [3.928710436279342e-08, 1.4040679026169526e-09, 6.297477869345885e-11, 6.0272729989784395e-09, 5.866597273975662e-07, 
#  4.621719265050956e-05, 0.001424895404519041, 0.010013652535366065, 0.027852813399730496, 0.04675124894441568]
# [0.00018054736134797994, 3.014122409645781e-05, 5.787138247597693e-06, 5.630261161955814e-05, 0.0005560835259970497, 
#  0.004985293618684871, 0.028784422261183794, 0.07677210662273688, 0.1264726843726221, 0.16511044354965607]

    # 构建数据集
    df_train,df_test = build_data(price_GE,step_k,MA_N)
    
    # 训练
    omega_list_j = []
    MSE_list_j = []
    MAE_list_j = []
    for i in range(5):
        row_start = batch_num * i
        row_end = batch_num * (i+1)
        df_train_verify = df_train[row_start:row_end] # (301,35)
        # print(df_train_verify.shape)
        df_train_train = pd.concat([df_train[0:row_start],df_train[row_end:]],axis=0) # (1204,35)
        # print(df_train_train.shape)
        omega_list_j.append(mytrain(df_train_train,lamb))
        mse,mae = myverify(df_train_verify,omega_list_j[i])
        MSE_list_j.append(mse)
        MAE_list_j.append(mae)
    omega_list.append(omega_list_j)
    MSE_list.append(MSE_list_j)
    mse_mean = sum(MSE_list_j)/len(MSE_list_j)
    mse_mean_list.append(mse_mean)
    MAE_list.append(MAE_list_j)
    mae_mean = sum(MAE_list_j)/len(MAE_list_j)
    mae_mean_list.append(mae_mean)
# print(omega_list[i])
# print(MSE_list)
# print('*'*20)
print(mse_mean_list)
print(mae_mean_list)
````

````python
# 绘图
def myplot(data_Y,pred_Y):
    length = data_Y.shape[0]
    x = np.arange(length)
    # 指定图像大小
    fig = plt.figure(figsize=(10,6))
    
    plt.plot(x, data_Y, 'b--', label='real')
    plt.plot(x, pred_Y, 'r:', label='predict')
    
    # 登记线系列
    plt.title("Predicted v.s. Real Opening Prices of GE 2016")
    plt.xlabel('trading day [2016-01-01,2016-12-31]')
    plt.ylabel('opening prices of GE')
    plt.grid(True)
    plt.legend(bbox_to_anchor=(1.0, 1), loc=1, borderaxespad=0.)
    plt.show()
````
