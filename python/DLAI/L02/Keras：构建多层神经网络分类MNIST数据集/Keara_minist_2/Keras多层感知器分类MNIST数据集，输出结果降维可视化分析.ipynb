```python
# 绘制像素热力图
def visualize_input(img, ax):
    # 先绘制数字的大图，然后对784个像素每一个像素标注灰度值
    ax.imshow(img, cmap='gray')
    width, height = img.shape
    thresh = img.max()/2.5
    for x in range(width):
        for y in range(height):
            ax.annotate(str(round(img[x][y],2)), xy=(y,x),
                        horizontalalignment='center',
                        verticalalignment='center',
                        color='white' if img[x][y]<thresh else 'black')
    

i = 7 # 绘制数据集中索引为7的图像
fig = plt.figure(figsize = (10,10)) # 设定图的总大小
ax = fig.add_subplot(111)
# 调用绘图函数
visualize_input(X_train[i], ax)
```

```python
# 绘制前16张图像，可以修改16这个数字，展示前若干张图像
for i in range(16):
    plt.style.use({'figure.figsize':(15,15)}) # 设置图像大小
    plt.subplot(1,4,i%4+1) # 绘制一行四列的子图
    plt.imshow(X_train[i]) # 绘制被分类错误的图片
    title = '真实标签:{}'.format(str(y_train[i])) # 图表标题说明图片的正确标签和错误结果
    plt.title(title)
    plt.xticks([]) # 关闭刻度线显示
    plt.yticks([])
    plt.axis('off') # 关闭坐标轴显示
    if i%4 == 3: # 凑够一行四个图就显示出来
        plt.show()
```

# 数据预处理：将图像展平成784维长向量、归一化、标签转独热向量编码

将长28宽28的图像拉平成784维长向量。

```python
X_train = X_train.reshape(60000, 784)
X_test = X_test.reshape(10000, 784)
X_train = X_train.astype("float32")
X_test = X_test.astype("float32")
```

```python
# 幅度缩放，归一化
X_train /= 255
X_test /= 255
```

```python
# 前十个训练集标签
y_train[:10]
```

原始数据的标签是用整数表示的，但在手写数字分类场景中，数字是9并不意味着比1大，也就是说标签之间是没有大小关系的，我们不能让神经网络处理这种关系，因此，我们需要将标签转化为独热向量编码。对应数字的索引为1。

One-Hot-Encoding

在构造损失函数时，用独热向量编码计算交叉熵损失函数。

```python
y_train[0]
```

```python
from keras.utils import np_utils
```

```python
# 把标签等转成one-hot的形式
Y_train = np_utils.to_categorical(y_train, 10)
Y_test = np_utils.to_categorical(y_test, 10)
```

# 构建多层感知机模型

```python
model = Sequential()
model.add(Dense(513, activation='relu', input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer=SGD(lr=0.001), \
              metrics=['accuracy'])
```

```python
# 配置Tensorboard可视化工具
from keras.callbacks import TensorBoard
import os
# 当前路径和上一级路径中不要包含中文
# logdir = os.join.path('cnn_selu_callbacks')
logdir = r'.\tensorboard\keras_log'
logdir = os.path.join('tensorboard_keras_log')
tensorboard = TensorBoard(log_dir=logdir, update_freq=1000)  
```

```python
# 训练25轮，每个batch输入128个数据
network_history = model.fit(X_train, Y_train, batch_size=128, \
                            epochs=50, verbose=1, validation_data=(X_val, Y_val),callbacks=[tensorboard])
```

# 可视化训练过程中误差变化

```python
import matplotlib.pyplot as plt
%matplotlib inline

def plot_history(network_history):
    plt.figure()
    plt.xlabel('训练轮次')
    plt.ylabel('损失函数')
    plt.plot(network_history.history['loss'])
    plt.plot(network_history.history['val_loss'])
    plt.legend(['训练集', '验证集'])
    plt.title('训练过程中损失函数变化')
    plt.grid()

    plt.figure()
    plt.xlabel('训练轮次')
    plt.ylabel('准确率')
    plt.plot(network_history.history['accuracy'])
    plt.plot(network_history.history['val_accuracy'])
    plt.legend(['训练集', '验证集'], loc='lower right')
    plt.title('训练过程中准确率变化')
    plt.grid()
    plt.show()

plot_history(network_history)
```

# 抽取512神经元的中间层输出结果并可视化


```python
model_truncated = Sequential()
model_truncated.add(Dense(513, activation='relu', input_shape=(784,)))
model_truncated.add(Dropout(0.2))
model_truncated.add(Dense(512, activation='relu'))

for i, layer in enumerate(model_truncated.layers):
    layer.set_weights(model.layers[i].get_weights())

model_truncated.compile(loss='categorical_crossentropy', optimizer=SGD(), 
              metrics=['accuracy'])
```

```python
model_truncated.summary()
```

```python
# 抽取特征
hidden_features = model_truncated.predict(X_train)
```

```python
hidden_features.shape
```

```python
hidden_features[:10000].shape
```

# 选取可视化的数字个数

```python
Number_Of_Show = 3000
```

```python
from sklearn.manifold import TSNE
# 使用T-SNE降维到2维，方便可视化
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(hidden_features[:Number_Of_Show])
```

```python
X_tsne
```

```python
# TSNE降维降到二维，散点图可视化
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'purple', 'coral', 'lime']
colors_map = np.argmax(Y_train, axis=1) # 将训练集标签的独热向量编码还原为整数
colors_map = colors_map[:Number_Of_Show] # 取前Number_Of_Show张图片
plt.figure(figsize=(10,10))
nb_classes = 10
for cl in range(nb_classes): # 遍历每一种数字
    indices = np.where(colors_map==cl) # 找到这种数字的所有图片的索引号
    plt.scatter(X_tsne[indices,0], X_tsne[indices, 1], c=colors[cl], label=cl, s=8)
plt.legend()
plt.show()
```

```python
# TSNE降维降到二维，散点图可视化，换颜色
# colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'purple', 'coral', 'lime']
colors_map = np.argmax(Y_train, axis=1)
colors_map = colors_map[:Number_Of_Show]
plt.figure(figsize=(10,10))
nb_classes = 10
for cl in range(nb_classes):
    indices = np.where(colors_map==cl)
    color = plt.cm.Set1(cl/10) # 自动生成新颜色
    plt.scatter(X_tsne[indices,0], X_tsne[indices, 1], color=color, label=cl, s=8)
plt.legend()
plt.show()
```

# 降维到三维

```python
Number_Of_Show = 1300
```

```python
from sklearn.manifold import TSNE
# 降维到3维，方便可视化
tsne3 = TSNE(n_components=3)
X_tsne3 = tsne3.fit_transform(hidden_features[:Number_Of_Show])
```

```python
# 将降维后的坐标归一化
x_min_3,x_max_3 = np.min(X_tsne3,0),np.max(X_tsne3,0)
X_tsne3 = (X_tsne3-x_min_3)/(x_max_3-x_min_3)
```

```python
# TSNE降维降到三维，散点图可视化
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 空间三维画图
%matplotlib inline

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'purple', 'coral', 'lime']
colors_map = np.argmax(Y_train, axis=1)
colors_map = colors_map[:Number_Of_Show] # 取前Number_Of_Show张图片
# 绘制散点图
fig = plt.figure(figsize=(15,10))
ax = Axes3D(fig)
for cl in range(nb_classes):
    indices = np.where(colors_map==cl)[0]
    ax.scatter(X_tsne3[indices[:120],0],X_tsne3[indices[:120],1],X_tsne3[indices[:120],2],c=colors[cl], label=cl)

# 添加坐标轴(顺序是Z, Y, X)
ax.set_zlabel('Z', fontdict={'size': 15, 'color': 'red'})
ax.set_ylabel('Y', fontdict={'size': 15, 'color': 'red'})
ax.set_xlabel('X', fontdict={'size': 15, 'color': 'red'})
plt.legend()
plt.show()
```

```python
# TSNE降维降到三维，散点图可视化 交互式
from mpl_toolkits import mplot3d
from ipywidgets import interact

def plot_3D(elev=30, azim=30):
    fig = plt.figure(figsize=(15,10))
    ax = fig.gca(projection='3d')
    
    for cl in range(nb_classes):
        indices = np.where(colors_map==cl)[0]
        ax.scatter(X_tsne3[indices[:120],0],X_tsne3[indices[:120],1],X_tsne3[indices[:120],2],c=colors[cl], label=cl)
    
        # ax.scatter3D(X_tsne3[:,0],X_tsne3[:,1],X_tsne3[:,2],c=y_train[:5000],cmap=plt.cm.Spectral,marker = 'o',s=50,label=y_train[:5000])
    ax.view_init(elev=elev, azim=azim)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    plt.legend()
    
interact(plot_3D, elev=[30,45,60],azim=(-180,180))
```
